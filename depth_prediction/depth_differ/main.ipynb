```python
import cv2
import json
import math
import os
import os.path
import re
import tempfile
import airsim
from matplotlib import pyplot
import quickstart as util
import utils_KITTI as utilkitti
```

```python
# 資料路徑
DATA_PATH = '/home/islab/Documents/drone/depth_prediction/data/random/data/2020-04-16-14-15/depth_planner/2020-04-16-14-21-27'
RESULT_PATH = '/home/islab/Documents/drone/depth_prediction/depth_differ/result/2020-04-16-14-21-27'

# Model Input 解析度設定
INPUT_HEIGHT = 320
INPUT_WIDTH = 384

# Video Output 解析度設定
OUTPUT_HEIGHT = 1440
OUTPUT_WIDTH = 1920
```

```python
result_image_folder_path = os.path.join(RESULT_PATH, 'depthPlanner')
if not os.path.exists(result_image_folder_path):
    os.makedirs(result_image_folder_path)
for i in range(49):
    # 讀取影像
    n = "%03d" % (i+1)
    m = "%03d" % (i+2)
    depth_before, scale = utilkitti.read_pfm(os.path.join(DATA_PATH, str(n)+'.pfm'))
    depth, scale = utilkitti.read_pfm(os.path.join(DATA_PATH, str(m)+'.pfm'))
    result = abs(depth - depth_before)
    # 儲存結果
    airsim.write_pfm(os.path.join(result_image_folder_path, str(n)+'.pfm'), result)
    print(os.path.join(result_image_folder_path, str(n)+'.pfm'))
    #plt.imsave(result_image_folder_path, result, cmap = plt.cm.gray)
#print('predicted dataset {}'.format(folder_name))
            
```

```python
result_image_folder_path = os.path.join(RESULT_PATH, 'depth_perspective')
if not os.path.exists(result_image_folder_path):
    os.makedirs(result_image_folder_path)
for i in range(49):
    # 讀取影像
    n = "%03d" % (i+1)
    m = "%03d" % (i+2)
    depth_before, scale = utilkitti.read_pfm(os.path.join(DATA_PATH.replace('depthPlanner','depth_perspective'), str(n)+'.pfm'))
    depth, scale = utilkitti.read_pfm(os.path.join(DATA_PATH.replace('depthPlanner','depth_perspective'), str(m)+'.pfm'))
    result = abs(depth - depth_before)
    # 儲存結果
    airsim.write_pfm(os.path.join(result_image_folder_path, str(n)+'.pfm'), result)
    print(os.path.join(result_image_folder_path, str(n)+'.pfm'))
```

```python
# prepare result video folder
result_video_folder_path = os.path.join(RESULT_PATH, 'video')
if not os.path.exists(result_video_folder_path):
    os.makedirs(result_video_folder_path)
# iterate result images
result_image_folder =  result_image_folder_path
# create result video
video_path = os.path.join(result_video_folder_path, '2019-12-05-05-59-58.mp4')
video = cv2.VideoWriter(video_path, cv2.VideoWriter_fourcc(*'X264'), 10, (OUTPUT_WIDTH, OUTPUT_HEIGHT))
# iterate each image
for file_name in sorted(os.listdir(os.path.join(result_image_folder)), key=lambda x: int(re.split('\\-|\\.', x)[-2])):
    # 儲存結果
    image_path = os.path.join(result_image_folder)
    image = cv2.resize(cv2.imread(image_path), (OUTPUT_WIDTH, OUTPUT_HEIGHT))
    video.write(image)
video.release()
print('{} created'.format(video_path))
```
